# Importamos librerías

```python
import tensorflow as tf
import pathlib
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import keras
```

# Primera prueba(Imagen propia)

## Importamos imagen de pureba

```python
#path1 = '../../../../Data/Processed/img_prop/test_me.jpg'
path1 = r'D:\Documentos de trabajo\Maestría\Curso Maestría Introducción a Redes Neuronales\Tarea 2\Tarea 2. Reconocimiento facial\Tareea2.-Curso-Redes-Neuronales.-Rafael-S-nchez-Cedillo\Data\Processed\img_prop\test_me.jpg'
```

```python
or_prueba1 = tf.keras.preprocessing.image.load_img(path1)
or_prueba1 = tf.image.resize(or_prueba1, [178,218])
or_prueba1 /= 255.0 
print(or_prueba1.shape)
print(or_prueba1.dtype)
```

```python
plt.imshow(or_prueba1)
```

```python
or_prueba1 =or_prueba1[None, :]
```

## Importamos el modelo y hacemos predicciones

```python
new_model = keras.models.load_model('FT.h5')
```

```python
new_predictions = new_model.predict(or_prueba1)
print(new_predictions)
```

```python
new_predictions = new_predictions.round()
```

```python
new_predictions
```

```python
if new_predictions== 1:
    print("No es Rafael")
else: 
    print("Es Rafael")
```
