```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../Data/Processed/atributes.csv')
df
```

```python
correlation_table= df.corr()
correlation_table
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(correlation_table,vmin=-1,vmax=1, cmap="jet_r")
plt.show()
```

```python
df_wd=df.drop(['Wearing_Necktie', 'Wearing_Necklace', 'Wearing_Lipstick', 
              'Wearing_Hat', 'Wearing_Earrings','Heavy_Makeup',
               'Attractive', 'Smiling'], axis=1)
```

```python
df_wd
```

```python
correlation_table2= df_wd.corr()
correlation_table2
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(correlation_table2,vmin=-1,vmax=1, cmap="jet_r")
plt.show()
```
